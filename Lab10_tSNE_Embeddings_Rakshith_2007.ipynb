```python
import gensim.downloader as api
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
model = api.load("glove-wiki-gigaword-100")   # 100-dimensional embeddings

print("Vocabulary Size:", len(model))

# Display example vector
print("\nVector for word 'king':\n")
print(model["king"])
```

```python
words = [
    # Animals
    "dog", "cat", "lion", "tiger", "wolf", "elephant", "horse", "monkey",

    # Countries
    "india", "china", "france", "germany", "japan", "america", "canada", "brazil",

    # Fruits
    "apple", "banana", "mango", "orange", "grape", "pineapple", "papaya", "cherry",

    # Technology
    "computer", "laptop", "keyboard", "mouse", "internet", "software", "hardware", "mobile",

    # Vehicles
    "car", "bus", "train", "bicycle", "airplane", "truck", "ship", "motorcycle",

    # Family
    "mother", "father", "brother", "sister", "child", "uncle", "aunt", "grandmother"
]

vectors = np.array([model[word] for word in words])
```

```python
tsne = TSNE(n_components=2, random_state=42, perplexity=5)
reduced_vectors = tsne.fit_transform(vectors)
```

```python
plt.figure(figsize=(12,10))

x = reduced_vectors[:, 0]
y = reduced_vectors[:, 1]

plt.scatter(x, y)

for i, word in enumerate(words):
    plt.annotate(word, (x[i], y[i]))

plt.title("t-SNE Visualization of Word Embeddings")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.show()
```

1️⃣ What problem does t-SNE solve?

 t-SNE reduces high-dimensional data (like 100 or 300 dimensions) into 2D or 3D for visualization.

2️⃣ Why can’t we directly visualize 300-dimensional vectors?

 Humans can only see 2D or 3D. 300D space cannot be plotted directly.

3️⃣ How does t-SNE help understand embeddings?

 It keeps similar words close together and separates different words.

4️⃣ Why do similar words appear together?

 Because their vectors are mathematically similar.

5️⃣ Why might unrelated words appear close?

 t-SNE sometimes preserves local structure better than global structure.

6️⃣ Two real-world uses:

* Detecting bias in embeddings

* Understanding semantic relationships in NLP models

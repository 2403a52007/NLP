```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
text = "Artificial Intelligence is transforming the world."

print(summarizer(text, max_length=30, min_length=10)[0]['summary_text'])
```

```python
texts = [
"""Artificial Intelligence is transforming industries by enabling machines to learn from data.""",

"""The Indian economy is growing with digital transformation and startups.""",

"""Natural Language Processing helps computers understand human language.""",

"""A farmer worked hard and finally achieved success after many struggles.""",

"""Clean code is important for better software development."""
]
```

```python
for i, text in enumerate(texts):
    summary = summarizer(text, max_length=50, min_length=20)

    print("\nTEXT", i+1)
    print(text)
    print("SUMMARY:", summary[0]['summary_text'])
```

Objective

The objective of this lab is to understand text summarization as a Sequence-to-Sequence task in Natural Language Processing and to generate summaries using pretrained models.

Dataset Description

The dataset consists of multiple text passages collected from sources like news articles, blogs, academic content, and short stories. Different types of texts were used to evaluate how the model performs on various writing styles and domains.

Methodology

A pretrained transformer-based summarization model was used to generate summaries. The model processes input text and produces a shorter version while maintaining important information. Different parameters were adjusted to control the length of summaries.

Results

The model successfully generated concise summaries from longer texts. It captured the main idea while removing unnecessary details. Short summaries were very brief, while longer summaries retained more information.

Analysis

The model performed well on structured texts like news and technical content. However, in narrative or conversational texts, some important details were lost. The summaries were generally coherent but sometimes missed context.

Conclusion

Text summarization using pretrained models is effective for reducing large text into concise information. However, some limitations such as loss of details and difficulty with complex texts were observed.

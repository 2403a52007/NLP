```python
import pandas as pd
import re
import nltk
from collections import defaultdict, Counter
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
data = pd.read_csv("/content/Twitter_Data.csv")
data = data[['clean_text']]  # column name may vary
data.head()
```

```python
def clean_tweet(text):
    text = str(text)   # FIX
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    text = re.sub(r"#\w+", "", text)
    text = re.sub(r"[^A-Za-z\s]", "", text)
    return text.lower().strip()
data['processed'] = data['clean_text'].apply(clean_tweet)
data.head()
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
tagged_sentences = []
for t in data['processed'].head(100):
    words = nltk.word_tokenize(t)
    tagged_sentences.append(nltk.pos_tag(words))
tagged_sentences[0]
```

```python
from collections import defaultdict, Counter
transition = defaultdict(Counter)
for sent in tagged_sentences:
    for i in range(len(sent)-1):
        transition[sent[i][1]][sent[i+1][1]] += 1
transition['NN']
```

```python
emission = defaultdict(Counter)
for sent in tagged_sentences:
    for word, tag in sent:
        emission[tag][word] += 1
emission['NN']
```

```python
from collections import Counter
word_count = Counter()

for sent in tagged_sentences:
    for word, tag in sent:
        word_count[word] += 1
rare_words = [w for w, c in word_count.items() if c == 1]
rare_words[:5]
```
